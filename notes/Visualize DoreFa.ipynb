```python
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
a = np.arange(-1, 1, 0.001)
```

```python
def quantize_k(r_i, k):
	scale = (2**k - 1)
	r_o = torch.round( scale * r_i ) / scale
	return r_o

k = 8
r_i = torch.from_numpy(a)
tanh = torch.tanh(r_i).float()
r_o = 2*quantize_k( tanh / (2*torch.max(torch.abs(tanh)).detach()) + 0.5 , k) - 1
```

```python
plt.plot(a, r_o.numpy())
plt.show()
```
